# 01. 노드 시작하기

## 서버란 무엇인가?

클라이언트가 필요한 정보나 서비스를 요청했을 때 해당 정보와 서비스를 제공해주는 것이 서버이다. 클라이언트는 요청을 보내는 주체로, 브라우저일 수도 있고 데스크톱, 모바일, 다른 서버로 요청을 보내는 서버일 수도 있다.

예를 들어 사용자가 네이버에 접근하고자 하면(요청), 브라우저는 네이버의 서버에 접근하여 브라우저에 렌더링 하기 위한 HTML/CSS 파일이라든지, javascript 파일 등을 가지고 클라이언트에 응답한다.

## 자바스크립트 런타임

노드는 자바스크립트 런타임이다. 런타임은 특정 언어로 만든 프로그램을 실행할 수 있는 환경을 뜻한다. 따라서 노드는 자바스크립트 프로그램을 컴퓨터에서 실행할 수 있게 해준다.

## 이벤트 기반

- 이벤트 기반이란 이벤트가 발생할 때 미리 지정해둔 작업을 수행하는 방식을 의미한다. 이벤트로는 클릭이나 네트워크 요청 등이 있을 수 있다.
- 이벤트 기반 시스템에서는 이벤트가 발생하면 어떤 일을 처리해야하는지 미리 등록해두어야 한다. 이것을 이벤트 리스너에 콜백 함수를 등록한다고 표현한다.

콜백 함수가 언제 호출 스택에 들어가는지 알기 위해 다음과 같은 세 가지 개념을 알아야 한다.
#### 1. 이벤트 루프
이벤트 발생 시 호출할 콜백 함수들을 관리하고, 호출된 콜백 함수의 실행 순서를 결정하는 역할을 담당한다. 노드가 종료될 때까지 이벤트 처리를 위한 작업을 반복하므로 루프라고 불린다.

#### 2. 태스크 큐
이벤트 발생 후 호출되어야 할 콜백 함수들이 기다리는 공간이다. 이벤트 루프가 정한 순서대로 줄을 서 있으므로 콜백 큐라고도 부른다.

#### 3. 백그라운드
타이머나 I/O 작업 콜백 또는 이벤트 리스너들이 대기하는 곳이다.

### 이벤트 발생 시 실행 순서
1. 호출된 함수는 호출 스택에 등록된다.
2. 가장 먼저 main() 함수가 호출 스택에 등록되고, 첫 줄부터 차례로 실행하며 각 함수가 호출 스택에 등록된다.
3. 다음으로 setTimeout()이 호출 스택에 들어간다. 스택은 FILO의 구조로 가장 나중에 들어간 setTimeout이 실행되고, 호출 스택에서 빠지게 된다. 이때 타이머와 함께 run 콜백을 백그라운드로 보낸다.
4. 3초가 지나면 run 콜백은 백그라운드에 의해 태스트 보내진다. 
5. 호출 스택 실행이 모두 끝나 비워지면 6번을 실행한다.
6. 이벤트 루프가 태스트 큐의 run 콜백을 호출 스택으로 올린다. 이때 호출 스택에 함수가 많이 차 있으면 3초가 지난 후에도 run 함수가 실행되지 않을 수도 있다. 따라서 setTimeout의 시간이 정확하지 않을 수도 있는 이유이다.

## 논블로킹 I/O

논블로킹이란 이전 작업이 완료될 때까지 기다리지 않고 다음 작업을 수행함을 뜻한다. 하지만 싱글 스레드라는 한계 때문에 자바스크립트의 모든 코드가 이 방법을 통한 시간적 이득을 볼 수 있는 것은 아니다. 현재 노드 프로세스 외의 다른 컴퓨팅 자원을 사용할 수 있는 I/O작업이 주로 시간적 이득을 많이 본다.

## 싱글 스레드

- 노드는 싱글스레드이다. 따라서 하나의 스레드만으로 모든 작업을 처리해야 한다.
- 노드는 싱글스레드, 논블로킹 방식을 채택하고 있다. 싱글스레드만으로 여러 일을 동시에 처리할 수 있다(병렬성이 아닌 동시성을 의미)는 장점이 있지만, 하나의 스레드가 제 역할을 하지 못하게 된다면 전체 시스템에 큰 문제가 생길 수 있다는 단점이 있다.
- 따라서 노드는 멀티프로세싱 방식을 택했다. 스레드가 아닌 프로세스 자체를 늘리는 것이다.
- 그 이유는 자바스크립트 언어 자체가 싱글 스레드 특성을 띠고 있기 때문이다.

#### 프로세스
프로그램이 실행중인 상태를 의미한다. 살아있는 프로그램! 운영체제에서 할당하는 작업의 단위이다. 프로세스 간에서는 주소공간을 절대 넘어가서는 안된다. 따라서 프로세스 간에는 메모리 등의 자원을 공유하지 않는다.

#### 스레드
프로세스 내에서 실행되는 흐름의 단위이다. 하나의 프로세스는 스레드를 여러개 가질 수 있으며, 스레드들은 모두 부모 프로세스의 자원을 공유한다.

## 서버로서의 노드

#### 노드를 서버로 사용하기에 적합한 프로그램
- 노드서버는 싱글 스레드, 논블로킹 방식을 채택했다. 따라서 노드서버는 I/O가 많은 작업에 적합하다.
- 개수는 많지만 크기는 작은 데이터를 실시간으로 주고받는 데 적합하다.
- 실시간 채팅 애플리케이션이나 주식차트, JSON 데이터를 제공하는 API 서버가 노드를 많이 사용한다.

#### 노드를 서버로 사용하기에 부적합한 프로그램
- CPU 연산을 많이 요구하는 프로그램에는 적합하지 않다는 의미이기도 하다. 코드가 CPU 연산을 많이 요구하면 블로킹이 발생해 스레드 하나가 감당하기 어렵다.
- 이미지나 비디오처리, 대규모 데이터 처리 같이 CPU를 많이 사용하는 작업을 위한 서버로는 권장되지 않는다.

#### 장점
1. 웹 서버가 내장(express)되어 있다. 따라서 Apache, nginx, IIS처럼 별도의 웹 서버를 설치할 필요 없어서 비교적 쉽게 입문할 수 있다.
